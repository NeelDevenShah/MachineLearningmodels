**This is an special type of the selector which is used to select the points i.e the upper left and the lower right from the given images and from the selected box we will get the data and the boxes are of the two types 'Text' and 'Checkbox'

-->One a rectangle is selected the user is asked to enter the type and the Name:

-->Type can be 'Text' and 'Checkbox'

-->Name can be anything

==>Procedure to use this program
1. Click on the upper left of an box and than the lower right
2. Make an input of the "type" i.e text or box
3. Than again make an input of the "name"
4. Than make follow the above process till we want the location of the box
5. Than at the ending to get the data click on the window screen and press 's', The output will be shown in the terminal, And take that output and save it in the main file as the array of roi
6. Close this program
C. The cpu can crash, So make the selection of the area caucently, Make sure you give some gap between each process, And if not required than make the new window minimized, It will help

```python
import cv2
import random

scale = 0.5
circles = []
counter = 0
counter2 = 0
point1 = []
point2 = []
myPoints = []
myColor = []

def mousePoints(event, x, y, flags, params):
    global counter, point1, point2, counter2, circles, myColor
    if event == cv2.EVENT_LBUTTONDOWN:
        if counter==0:
            point1 = int(x//scale),int(y//scale);
            counter +=1
            myColor = (random.randint(0, 2)*200, random.randint(0, 2)*200, random.randint(0, 2)*200)
        elif counter ==1:
            point2 = int(x//scale), int(y//scale)
            type = input('Enter Type: ')
            name = input('Enter Name: ')
            myPoints.append([point1, point2, type, name])
            counter = 0
        circles.append([x, y, myColor])
        counter2 +=1
        
img = cv2.imread('formData/blankForm.png')
img = cv2.resize(img, (0, 0), None, scale, scale)

while True:
    # To Display Points
    for x, y, color in circles:
        cv2.circle(img, (x, y), 3, color, cv2.FILLED)
    cv2.imshow("Original Image", img)
    cv2.setMouseCallback("Original Image", mousePoints)
    if cv2.waitKey(1) & 0xFF == ord('s'):
        print(myPoints)
        break
```
