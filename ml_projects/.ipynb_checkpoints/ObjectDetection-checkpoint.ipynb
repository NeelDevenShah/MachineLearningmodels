```python
import cv2
import numpy as np

def nothing(x):
    pass

cap = cv2.VideoCapture(0)

# Creating the window for showing the images
cv2.namedWindow("parametres")
# Creating the trackbar for the threshold 1
                                              #t1 Here by checking(072)
cv2.createTrackbar("Threshold1", "parametres", 72, 255, nothing)

# Creating the trackbar for the threshold 2
                                              #t2 Here by checking(206)
cv2.createTrackbar("Threshold2", "parametres", 206, 255, nothing)

def getContours(img, imgContour):   
    # For detecting the contours    
    contours, hierarchy = cv2.findContours(img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    
    # for cnt in contours:
        # For drawing the contours 
        # cv2.drawContours(imgContour, contours, -1, (255, 0, 255), 7)
        
    #Now as the disturbance is comming while making the contour and extra contours are maked due to it, to solve it we will make an condition that if the area of a particular shape is more than the specified one than only the contour itto be made otherwise not
    for cnt in contours:
        area = cv2.contourArea(cnt)
        if area > 1000:
            cv2.drawContours(imgContour, cnt, -1, (255, 0, 255), 7)
            # Till now, we have focused on the making the contour around the object, but now we will make an box around the object
            # And for that we will get the dimensions of an box if made around the object
                
            # For getting the perimeter of the contour formed
            peri = cv2.arcLength(cnt, True)
                
            # For shape approximation from the given data
            approx = cv2.approxPolyDP(cnt, 0.02*peri, True)
            
            # print(len(approx))
            x_, y_, w, h = cv2.boundingRect(approx)
            # Making the outer rectangle
            cv2.rectangle(imgContour, (x_,y_), (x_+w,y_+h), (0, 255, 0), 5)
                
            # Printing the information about the object
            cv2.putText(imgContour, "Points: "+str(len(approx)), (x_ + w + 20, y_ + 20), cv2.FONT_HERSHEY_COMPLEX, 0.7, (0, 255, 0), 2)
            cv2.putText(imgContour, "Area: "+str(int(area)), (x_ + w + 20, y_ + 45), cv2.FONT_HERSHEY_COMPLEX, 0.7, (0, 255, 0), 2)
            
while(True):
    # Reading the image from the vedio
    success, img = cap.read()
    
    # Storing the value of the threshold1 for each frame    
    threshold1 = cv2.getTrackbarPos("Threshold1", "parametres")
    
    # Storing the value of the threshold2 for each frame
    threshold2 = cv2.getTrackbarPos("Threshold2", "parametres")
    
    # Making the image to be blur    
    imgBlur = cv2.GaussianBlur(img, (7, 7), 1)
    
    # Making the image to be gray
    imgGray = cv2.cvtColor(imgBlur, cv2.COLOR_BGR2GRAY)
    
    # Applying the canny edge detetion algorithm
    
    # Step 1:    
    # By finding the good values of the threshold1, and threshold2 by changing and experimenting the
    # different traceback values finding one and set them initially at the top two parameters
    imgCanny = cv2.Canny(imgGray, threshold1, threshold2)
    
    # Step 2:    
    #After selecting the perfect value for the threshold 1 and 2, we will make the image to be dilate by help of the kernel(mask)
    kernel = np.ones((5, 5))
    imgDil = cv2.dilate(imgCanny, kernel, iterations=1)
    
    # Step 3: Making the contours and maximum area contours(by help of points) tobe seen in the image and also printing the data of it in image...
    # The process of the step 3 is been caried out by the getCountours method buit in the above code
    
    # Making an image to be copy of initial image to show the contors in the pure un-processed image
    imgContour = img.copy()
    
    # Storing the image with contour in imgContour    
    getContours(imgDil, imgContour)
         
    
    # Showing the result on the screen    
    cv2.imshow('parametres', imgContour)
    # Codition for stoping the vedio window and it can be stopped by clicking 'q'    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
