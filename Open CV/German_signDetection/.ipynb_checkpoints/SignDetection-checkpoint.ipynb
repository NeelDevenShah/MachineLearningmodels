**Signal Sign detection by help of the german sign data by help of tensorflow, keras, opencv, pickle, matplotlib, pandas

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from keras.utils.np_utils import to_categorical
from keras.layers import Dropout, Flatten
from keras.layers.convolutional import Conv2D, MaxPooling2D
import cv2
from sklearn.model_selection import train_test_split
import pickle
import os
import pandas as pd
import random
from keras.preprocessing.image import ImageDataGenerator
```

```python
#General Parameters to be used in the program

# Folder with all the class folders
path="myData"

# file with all names of class
# labelFile = 'labels.csv'

# how many to process together
batch_size_val = 50

# Number of the cycles to be performed for each images
epoch_val = 10
imageDimension = (32, 32, 3)

# If 1000 images split will give 200 for testing
testRatio = 0.2
# If 1000 images 20% of remaining 800 will be 160 for the validation
validationRatio = 0.2
```

```python
# Importing the images
count =0
images = []
classNo = []
# For getting the folder information inside the myData folder
myList = os.listdir(path)
print("Total Classes Detected: ", len(myList))
```

```python
noOfClass = len(myList)
print("Importing the classes.....")
for x in range(0, len(myList)):
    myPictList = os.listdir(path+"/"+str(count))
    for y in myPictList:
        curImg = cv2.imread(path+"/"+str(count)+"/"+y)
        images.append(curImg)
        classNo.append(count)
    print(count, end=" ")
    count+=1
print(" ")
images = np.array(images)
classNo = np.array(classNo)
```

```python
# Spliting the data
x_train, x_test, y_train, y_test = train_test_split(images, classNo, test_size=testRatio)
x_train, x_validation, y_train, y_validation = train_test_split(x_train, y_train, test_size=validationRatio)

# Here,
# x_train = Array of images to be trained
# y_train = Corresponding class Id
```

```python
# To check if the number of images matches to number of labels for each data set
print("Data Shapes")
print("Train", end="")
print(x_train.shape, y_train.shape)
print("Validation", end="")
print(x_validation.shape, y_validation.shape)

# The assert keyword is used when debugging code. The assert keyword lets you test if a condition in your code returns True, if not, the program will raise an AssertionError.

assert(x_train.shape[0]==y_train.shape[0])
assert(x_validation.shape[0]==y_validation.shape[0])
assert(x_test.shape[0]==y_test.shape[0])
assert(x_train.shape[1:]==(imageDimension))
assert(x_validation.shape[1:]==(imageDimension))
assert(x_test.shape[1:]==(imageDimension))
```

```python
# Readind the csv file of the labels
data = pd.read_csv('labels.csv')
print("data shape", data.shape, type(data))
```

```python
# Display some sample images fo all the class
num_of_samples = []
cols = 5
num_classes = noOfClass
fig, axs = plt.subplots(nrows=num_classes, ncols=cols, figsize=(5, 300))
fig.tight_layout()
for i in range(cols):
    for j,row in data.iterrows():
        x_selected = x_train[y_train == j]
        axs[j][i].imshow(x_selected[random.randint(0, len(x_selected)- 1), :, :], cmap=plt.get_cmap("gray"))
        axs[j][i].axis("off")
        if i == 2:
            axs[j][i].set_title(str(j)+ "-"+row["Name"])
            num_of_samples.append(len(x_selected))
```

```python
# Processing the images
def grayscale(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return img

def equalize(img):
    img = cv2.equalizeHist(img)
    return img

def preprocesssing(img):
    img = grayscale(img) # Convert to GRAYSCALE
    img = equalize(img) # Standardize the lighting in an image
    img = img/255 # To normal values between 0 and 1 instead of 0 to 255
    return img

x_train = np.array(list(map(preprocesssing, x_train))) # To iterate and preprocess all images
x_validation = np.array(list(map(preprocesssing, x_validation)))
x_test = np.array(list(map(preprocesssing, x_test)))
```

```python
# Add a depth of 1
x_train = x_train.reshape(x_train.shape[0], x_train.shape[1], x_train.shape[2], 1)
x_validation = x_validation.reshape(x_validation.shape[0], x_validation.shape[1], x_validation.shape[2], 1)
x_test = x_test.reshape(x_test.shape[0], x_test.shape[1], x_test.shape[2], 1)
```

```python
# Argumentation of images: To makeit more generic
dataGen = ImageDataGenerator(width_shift_range=0.1, height_shift_range=0.1, zoom_range=0.2, shear_range=0.1, rotation_range=10)
dataGen.fit(x_train)
# Requesting data generator to generate images of batch size = no. of images creaed each time its called
batches = dataGen.flow(x_train, y_train, batch_size=20)
x_batch, y_batch = next(batches)
```

```python
# Converting a class vector to binary class matrix

# to_categorical converts a class vector into a binary class matrix
y_train = to_categorical(y_train, noOfClass)
y_validation = to_categorical(y_validation, noOfClass)
y_test = to_categorical(y_test, noOfClass)
```

```python
# Convolution Neural Network Model
def myModel():
    no_of_Filters = 60
    
    # This is the kernel that moves around the image to get the features    
    size_of_Filter=(5,5)
    
    #This would remove 2 pixels from each border when using 32 32 image
    size_of_Filter2=(3,3) 
    
    # Scale down all feature map to generalize more, to reduce overfitting
    size_of_pool = (2,2)
    
    # Number of nodes in hidden layer
    no_Of_Nodes = 500
    
    model = Sequential()
    # Adding more convolution layer = less features but can cause accuracy to increase
    model.add((Conv2D(no_of_Filters, size_of_Filter, input_shape=(imageDimension[0], imageDimension[1], 1), activation='relu')))
    model.add((Conv2D(no_of_Filters, size_of_Filter, activation='relu')))
    # Does not affect the depth/no. of filters
    model.add(MaxPooling2D(pool_size=size_of_pool))
    
    model.add((Conv2D(no_of_Filters//2, size_of_Filter2, activation='relu')))
    model.add((Conv2D(no_of_Filters//2, size_of_Filter2, activation='relu')))
    model.add(MaxPooling2D(pool_size=size_of_pool))
    model.add(Dropout(0.5))
    
    model.add(Flatten())
    model.add(Dense(no_Of_Nodes, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(noOfClass, activation='softmax')) #Output layer
    
    #Compiling the model
    model.compile(Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])
    return model
```

```python
# Training the model
model = myModel()
print(model.summary())
```

```python
history = model.fit(dataGen.flow(x_train, y_train, batch_size=batch_size_val), steps_per_epoch=(len(x_train) // batch_size_val), epochs=epoch_val, validation_data=(x_validation, y_validation), shuffle=1)
```
