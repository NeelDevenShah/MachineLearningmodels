**Document scanner and saver by clicking the 's' for getting the scanned document

The utility code for this program is stored in the other file in the same folder by name of the utilities.py and which is imported in this file

Main code for making the document to be scanned

```python
import cv2
import numpy as np
import utilities

# Configuring the web-cam
cap = cv2.VideoCapture(0)

def nothing(x):
    pass

cv2.namedWindow("Result")
cv2.resizeWindow("Trackbars", 360, 240)
cv2.createTrackbar("Threshold1", "Result", 23,255, nothing)
cv2.createTrackbar("Threshold2", "Result", 43, 255, nothing)
count=0
    
while True:
    ret, img = cap.read()
    
    # Converting the image into the gray scale
    imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # Add gaussian blur
    imgBlur = cv2.GaussianBlur(imgGray, (5,5),1)
    
    th1 = cv2.getTrackbarPos("Threshold1", "Result")
    th2 = cv2.getTrackbarPos("Threshold2", "Result")
    imgThreshold = cv2.Canny(imgBlur, th1, th2)
    kernel = np.ones((5,5))
    
    # Applying the dilation
    imgDial = cv2.dilate(imgThreshold, kernel, iterations=2)
    
    # Applying the erosion
    imgThreshold = cv2.erode(imgDial, kernel, iterations=1)
    
    # Find all the contours
    imgContors = img.copy()
    imgBigContour = img.copy()
    
    # For finding all the contours
    contours, hierarchy = cv2.findContours(imgThreshold, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    # For drawing the contours
    cv2.drawContours(imgContors, contours, -1, (0,255,0), 10)
    
    # Process for finding the biggest contour and making it to be shown
    biggest, maxArea = utilities.biggestContour(contours)
    
    if biggest.size != 0:
        # Makking the data of the contour be in order
        biggest=utilities.reorder(biggest)
        # Drawing the biggest contour        
        cv2.drawContours(imgBigContour, biggest, -1, (0, 255, 0), 20)
        imgBigContour = utilities.drawRectangle(imgBigContour, biggest, 2)
        
        # Now making the image to be wrapped, i.e. croping the image such that the only data will be there that we want
        # Preparing the points for the wrap
        pts1 = np.float32(biggest)
        pts2 = np.float32([[0, 0], [480, 0], [0, 640], [480, 640]])
        matrix = cv2.getPerspectiveTransform(pts1, pts2)
        imgWarpColored = cv2.warpPerspective(img, matrix, (480, 640))
        
        # Removing 20 pixels from each sides
        imgWarpColored=imgWarpColored[20:imgWarpColored.shape[0] - 20, 20:imgWarpColored.shape[1] - 20]
        imgWarpColored = cv2.resize(imgWarpColored,(480,640))
        
        # Applying Adaptive threshold
        imgWarpGray = cv2.cvtColor(imgWarpColored, cv2.COLOR_BGR2GRAY)
        imgAdaptiveThre = cv2.adaptiveThreshold(imgWarpGray, 255, 1, 1, 7, 2)
        imgAdaptiveThre = cv2.bitwise_not(imgAdaptiveThre)
        imgAdaptiveThre=cv2.medianBlur(imgAdaptiveThre, 3)
        
        # stackedImage = utilities.stackImages([img, imgThreshold, imgContors, imgBigContour], 0.75)
        cv2.imshow("Result", imgAdaptiveThre)
    else:
        stackedImage = utilities.stackImages([img, imgThreshold, imgContors, img], 0.75)
        cv2.imshow("Result", stackedImage)
    #For making to show the results
    
    # For stoping the loop of the vedio and scanning
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break;
        
    # Code for saving the image on clicking the 's'
    if cv2.waitKey(1) & 0xFF == ord('s'):
        cv2.imwrite("Scanned/myImage"+str(count)+".jpg",imgWarpColored)
        cv2.putText(stackedImage, "Scan Saved")
        cv2.putText(imgAdaptiveThre, "Scan Saved")
        cv2.waitKey(300)
        count+=1
        
cap.release()
cv2.destroyAllWindows()
```
