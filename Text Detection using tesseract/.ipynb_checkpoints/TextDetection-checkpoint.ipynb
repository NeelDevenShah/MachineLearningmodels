**Text Detection With Opencv and tesseract
-->For this we have installed the two libraries
1. tesseract
2. pytesseract

```python
import cv2
import pytesseract
# Load an image
img = cv2.imread("sample3.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# For Printing the data that is been identified
print(pytesseract.image_to_string(img))
```

```python
# For getting the location for the each identified character with that particular character
Locationdata = pytesseract.image_to_boxes(img)
print(Locationdata)
```

```python
# For making the above location data in the formated way and make that particular alphabet to be highlited and under that we will print the value of that particular alphabet
imgHeight, imgWidth, _ = img.shape
boxes = pytesseract.image_to_boxes(img)
for b in boxes.splitlines():
    print(b)
    b = b.split(' ')
    print(b)
    x, y, w, h = int(b[1]), int(b[2]), int(b[3]), int(b[4])
    # (Initial points),(Initial points + width/height), color
    cv2.rectangle(img, (x,imgHeight-y), (w, imgHeight-h),(0,0,255),3)
    cv2.putText(img, b[0], (x, imgHeight-y+25), cv2.FONT_HERSHEY_COMPLEX, 1, (50,50,255),2)

cv2.imshow("Result", img)
# To load and hold the image, And if zero than it will close the window when the any key is pressed from the keyboard
cv2.waitKey(0)
# To close the window after the required jill was produced
cv2.destroyAllWindows()
```

```python
# For making to detect the whole word of the image
img =cv2.imread('sample3.jpg')
imgHeight, imgWidth, _ = img.shape
boxes = pytesseract.image_to_data(img)
print(boxes)
print("---------------")

# Sample code for example output
for x, b in enumerate(boxes.splitlines()):
    if x!=0:
        b = b.split()
        print(b)
    # Here from the above data we can see that if word is detected by the tesseract than only there would be the 12th index of the tuple otherwise not, Hence by help of this information we can write our further code

# Original code for output
for x, b in enumerate(boxes.splitlines()):
    if x!=0:
        b = b.split()
        if len(b) == 12:
            x, y, w, h = int(b[6]), int(b[7]), int(b[8]), int(b[9])
            cv2.rectangle(img, (x, y), (w+x, h+y), (0, 0, 255), 3)
            cv2.putText(img, b[11], (x,y), cv2.FONT_HERSHEY_COMPLEX, 1, (50, 50, 255), 2)
cv2.imshow("Result", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Method for detecting only numbers by help of the tesseract
# And for it we have to add some of the configuration to the pytesseract as shown in the figure below
img = cv2.imread('sample3.jpg')
imgHeight, imgWidth, _ = img.shape
config = r'--oem 3 --psm 6 outputbase digits'
boxes = pytesseract.image_to_boxes(img, config=config)
for b in boxes.splitlines():
    b = b.split(' ')
    print(b)
    x, y, w, h = int(b[1]), int(b[2]), int(b[3]), int(b[4])
    cv2.rectangle(img, (x, imgHeight-y), (w, imgHeight-h), (0, 0, 255), 3)
    cv2.putText(img, b[0], (x, imgHeight-y+25), cv2.FONT_HERSHEY_COMPLEX, 1, (50, 50, 255), 2)
cv2.imshow("Result", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
# The changes in the configuration can be made like oem and psm, as per the documentation refer it, screen shorts for the method of oem and psm is attached in the folder
```
